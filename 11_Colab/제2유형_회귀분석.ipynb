```python
import numpy as np
import pandas as pd
import seaborn as sns

bike = pd.read_csv('hour.csv')
bike.head(5)
```

```python
# 컬럼 확인
bike.columns
```

```python
bike.info()
```

```python
bike.describe()
```

```python
bike = bike.drop(['instant'], axis=1)
bike.head(5)
```

```python
# 날짜 인덱스 제거
bike = bike.drop(['dteday'], axis = 1)
bike.head(5)
```

```python
bike = bike.drop(['casual', 'registered'], axis = 1)
bike.head(5)
```

```python
category_list = ['season', 'hr', 'weekday', 'weathersit', 'workingday']

# 백업 데이터 생성
bike_dummy = bike.copy()

for category in category_list :
  dummies = pd.get_dummies(bike_dummy[category], prefix = category, drop_first = False)
  bike_dummy = pd.concat([bike_dummy, dummies], axis=1)

print(sorted(bike_dummy.columns.values))
```

```python
from sklearn.model_selection import train_test_split

y_target_d = bike_dummy['cnt']
X_feature_d = bike_dummy.drop(['cnt'], axis=1, inplace=False)

Xtrain, Xtest, ytrain, ytest = train_test_split(X_feature_d, y_target_d, test_size = 0.3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()
lr.fit(Xtrain, ytrain)
y_pred = lr.predict(Xtest)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(ytest, y_pred)

rmse = np.sqrt(mse)
r2 = r2_score(ytest, y_pred)
```

```python
print(mse, rmse, r2)
```
