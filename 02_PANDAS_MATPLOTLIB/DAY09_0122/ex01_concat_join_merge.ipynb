### [ 병합 - (2) merge ]
- 두 개의 DF에서 특정 컬럼을 기준으로 데이터를 병합하는 방식.
- 두 DF에 합치는 기준이 되는 컬럼을 지정해야 함.
- 지정된 컬럼명이 없으면 동일한 컬럼명을 기준으로 병합.

- (2-1) 데이터 준비

```python
f1 = '../DATA/stock price.xlsx'
f2 = '../DATA/stock valuation.xlsx'
```

```python
df1 = pd.read_excel(f1)
df2 = pd.read_excel(f2)
```

-(2-3) 데이터 확인

```python
df2.info()
```

(2-4) 데이터프레임 병합

- (2-4-1) merge() 

```python
# [기본] 두 개의 DF에서 동일 컬럼명을 찾아서 두 DF에 존재하는 컬럼명의 행 데이터만 병합.
pd.merge(df1, df2)
```

```python
df1.shape, df2.shape
```

```python
# [설정 => how 'outer'] 두 개의 DF에서 동일 컬럼명을 찾아서 모두 합치기.
pd.merge(df1, df2, how ='outer')
```

```python
# [설정 => how 'outer'] 두 개의 DF에서 동일 컬럼명을 찾아서 모두 합치기.
pd.merge(df1, df2, how ='outer')
```

```python
# [설정 => how 'left'] 두 개의 DF에서 동일 컬럼명을 찾아서 왼쪽 데이터를 기준으로 병합.
pd.merge(df1, df2, how ='left')
```

```python
# [설정 => how 'left'] 두 개의 DF에서 동일 컬럼명을 찾아서 오른쪽 데이터를 기준으로 병합.
pd.merge(df1, df2, how ='right') #df2에만 있었던 것은 제거됨.
```

```python
# [설정 => how 'left'] 두 개의 DF에서 동일 컬럼명을 찾아서 오른쪽 데이터를 기준으로 병합.
pd.merge(df1, df2, how ='cross') # how = 'cross' ----> 'df1의 0번 행과 df2의 모든 행, df2의 1번 행과 df2의 모든 행....'과 같은 방식으로 늘어남.
# 멀티인덱스..!
```
