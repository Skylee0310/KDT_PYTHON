```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v
df = pd.DataFrame([(.2, .3), (.0, .6), (.6, .0), (.2, .1)],
                  columns=['dogs', 'cats'])
df
```

```python
df.corr(method=histogram_intersection)
```

```python
df.corr()          # 개와 고양이는 음의 상관관계에 있다.
```

```python
def scatterplot(sr1, sr2):
    '''
    parameters: (파라미터가 무엇인지 독스트링으로 달아준다)
        sr1, sr2 : Series object
    '''
    plt.scatter(sr1, sr2)
    plt.title(f'{sr1.name} - {sr2.name} [{round(corrDF[sr1.name][sr1.name],2)}]')
    plt.xlabel(sr1.name)
    plt.ylabel(sr2.name)
    plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True) # y축 공유 / sharex = True ---> x축 공유
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar")) #proj-- 없으면 그냥 선그래프 됨.
#fig, axs = plt.subplots(2, 2) --> 이거만 실행시키면 그래프에 값이 출력되지 않음.
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
```

```python
# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)

# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
```
