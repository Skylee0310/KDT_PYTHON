```python
import pandas as pd
import numpy as np

X_train = pd.read_csv('X_train_mushroom.csv')
X_test = pd.read_csv('X_test_mushroom.csv')
y_train = pd.read_csv('y_train_mushroom.csv')
```

```python
X_train.head(5)
```

```python
X_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for col in X_train.columns:
  X_train[col] = label_encoder.fit_transform(X_train[col])
  X_test[col] = label_encoder.fit_transform(X_test[col])
```

```python
X_train.info()
```

```python
y_train = pd.get_dummies(y_train['class'])
y_train
```

```python
from sklearn.model_selection import train_test_split
# 훈련데이터와 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state = 42)
```

```python
print('X_train shape : ', X_train.shape)
print('X_test shape : ', X_test.shape)
print('y_train shape : ', y_train.shape)
print('y_test shape : ', y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

mdl = RandomForestClassifier(n_estimators=100, random_state = 42)
mdl.fit(X_train, y_train)

y_pred = mdl.predict(X_test)[:1300]
```

```python
from sklearn.metrics import accuracy_score, f1_score

f1 = f1_score(y_test.values.argmax(axis=1), y_pred.argmax(axis=1), average='weighted')
print(f1)
```
