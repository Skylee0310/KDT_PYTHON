### Softmax(), sigmoid() 함수 <hr> 
- sigmoid() : y=wx+b 결과 1개 전달 받음 ===> 반환 : 0.0 ~ 1.0 사이 값으로 변환 => 이진분류
- softmax() : y=wx+b 결과 n개 전달 받음 => 반환 : 여러 개의 결과 값의 합이 1.0이 되도록 변환 => 다중 분류 

```python
# 모듈 로딩
import torch                    # 토치 기본 모듈. Tensor 및 기본 함수들.
import torch.nn.functional as F # 파이토치 인공신경망 관련 함수들
```

```python
# 임의의 데이터 생성
data1 = torch.tensor([1,2,1,1,2])
data1.shape
```

```python
F.sigmoid(data1)
# 각 값을 0에서 1사이의 숫자로 조정.
```

- softmax()

```python
# softmax(
#     input: Tensor,
#     dim: int | None = ...,
#     _stacklevel: int = ...,
#     dtype: _dtype | None = ...
# ) -> Tensor
```

```python
# 임의의 데이터
data2 = data1.unsqueeze(dim=0) # 1행 5열
data2.shape, data2
```

```python
# 랜덤시드 설정
torch.manual_seed(2)
#
input = torch.randn(2,3) #torch.randn : 평균이 0이고 표준편차가 1인 정규분포(표준 정규분포)에서 무작위 생성.
output = F.softmax(input) # 0과 1 사이의 값으로 정규화
input, output 
```

```python
output.sum(dim=1) #각 행의 합
```

```python
output.max(dim=1) # values 와 indices 출력
output.max(dim=1)[1] # indices 출력
output.max(dim=1)[1][0] # 첫번째 indices 출력
output.max(dim=1)[1][0].item() # 첫번째 indices 출력
```

```python
output.argmax(dim=1) #최댓값을 가지는 요소의 인덱스를 반환
```
