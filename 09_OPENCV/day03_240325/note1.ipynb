아이리스 데이터를 가지고 DNN 모델을 다중분류로 설계할때
- 데이터 구조 : 4개 + 라벨 1개 (=> 분류 클래스 개수 : 3개)
    * 입력층 : Linear( 피처의 개수(4), 원하는 값(12)) 
        -  w의 개수 4*12 / b의 개수 12개
    * 은닉층 : Linear(12, 8) 
        - w의 개수 12 * 8 / b의 개수 8개
    * 출력층 : Linear(8, 3) # versicolor, verginica, setosa   
        * w의 개수 8*3 / b의 개수 3개

아이리스 데이터를 가지고 DNN 모델을 이진분류로 설계할때


- 데이터 구조 : 4개 + 라벨 1개 (=> 분류 클래스 개수 : 1개 => setosa와 그외)
    * 입력층 : Linear( 피처의 개수(4), 원하는 값(12)) # 퍼셉
    * 은닉층 : Linear(12, 8)
    * 출력층 : Linear(8, 1) # setosa
* 이제 이해했음. 다중분류일때는 하나하나하나가 다 클래스인데 이진분류일때는 'Setosa/그게 아닌 거!' 라서 마지막에 1개인것.

- 회귀일 경우 AF 필요 없음

- 입력 데이터의 shape(bs, F) 

shape(1, 3, 30, 30) =  (사진수, 채널, 높이, 너비)

* nn.Conv2D 클래스
    - 커널 1개 => w = (3*3=) 9개 / b = 1 (커널 10개 => 90개, b => 10개)
