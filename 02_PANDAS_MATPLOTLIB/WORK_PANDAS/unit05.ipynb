5.5 중복처리

5.5.1 duplicated 메서드

```python
empDF['Team'].head()
```

```python
empDF['Team'].duplicated().head() #duplicated()는 어떤 값이 이전에 나온 적이 있을 때 True를 반환. 처음 나오는 값은 False.
```

```python
empDF['Team'].duplicated(keep = 'first').head() #first는 기본 인수. 중복값이 처음 나타난 인덱스를 False 표시.
```

```python
empDF['Team'].duplicated(keep = 'last').head() #last는 열에서 마지막으로 나타나는 값을 중복되지 않은 것으로 표시.
```

```python
#각 팀에서 한 명의 직원을 추출. Team열에서 각 고유한 팀에 대한 첫번째 행을 가져올 수 있음.
# 첫번째 나온값은 원래 False이지만 반전시켜 첫번째값만 True인 값을 구할 수 있다.
```

```python
(~empDF['Team'].duplicated(keep = 'first')).head()
```

```python
firstOneInTeam = ~empDF['Team'].duplicated()
empDF[firstOneInTeam] #[첫번째 값만 True]인 인덱스의 값을 불러옴.
```

5.5.2 drop_duplicates 메서드
- 기본적으로 행의 모든 값이 일치하는 행을 제거.

```python
empDF.drop_duplicates()
```

```python
empDF.drop_duplicates(subset=['Team']) #Team에서 중복값을 모두 삭제.
```

```python
empDF.drop_duplicates(subset=['Team'], keep = 'last')
```

```python
empDF.drop_duplicates(subset=['First Name'], keep = False) #False를 인수로 전달하면 중복값이 있는 모든 행을 제거할 수 있음. 
```

```python
# 이름이 더글라스이고 남자인 모든 직원 추출
ManDouglas = (empDF['Gender'] == 'Male') & (empDF['First Name'] == 'Douglas')
empDF[ManDouglas]
```

```python
empDF.drop_duplicates(subset=['Gender', 'Team']).head()
# 인덱스 0의 행은 empDF에서 이름이 Douglas이고 성별이 Male인 직원.
# 판다스는 이름과 성별값이 인덱스 0과 동일한 나머지행을 삭제.
# 이름이 Douglas이지만 성별이 male이 아닌 경우와 male이지만 이름이 Douglas가 아닌 경우는 유지.
```
