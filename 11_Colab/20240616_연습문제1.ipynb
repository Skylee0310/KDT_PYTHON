```python
# 문제 2 - Age의 결측값을 평균으로 채우고 Age를 표준정규분포로 스케일링한 값을 Age_scale 열에 저장하시오. 답안은 print(df['Age_scale'])의 형태로 출력하시오.
```

```python
# 문제 4 - 생존한 사람과 생존하지 않은 사람들의 나이의 평균값을 순서대로 소수점 둘째 자리까지 구하시오. 단, 나이의 결측값은 전체 탑승자의 평균으로 채우시오.
```

```python
df['age'].isna().sum()
```

```python
survived = df[df['survived']==1]
not_survived = df[df['survived']==0]
```

```python
survived_mean = survived['age'].mean()
not_survived_mean = not_survived['age'].mean()
```

```python
print('생존자 나이 평균값 : ', survived_mean.round(2), '\n사망자 나이 평균값 : ', not_survived_mean.round(2))
```

```python
print('생존자 나이 평균값 : ', round(survived_mean, 2), '\n사망자 나이 평균값 : ', round(not_survived_mean, 2))
```
