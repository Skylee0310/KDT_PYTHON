### Tensor 원소/요소 접근 
- 인덱스 슬라이싱

```python
# 모듈 로딩
import torch
```

```python
# 텐서 생성
points = torch.tensor([[4., 2.], [5., 3.], [6., 4.]], dtype = torch.int16)
points
# DeprecationWarning
```

```python
# 텐서의 원소 1개 접근
# 데이터 3 추출
print(points[1,1])
```

```python
# 데이터 1번 행부터 나머지 모든 행 추출
print(points[1:])

# 데이터 4 추출
print(points[-1, -1])
print(points[2, 1])
```

브로드 캐스팅 연산
- 다른 모양(shape)을 가진 텐서들 간에 산술 연산을 수행할 때 발생하는 자동 변환 과정

```python
m1 = torch.FloatTensor([[1,2]]) # 실수 텐서 생성자
m2 = torch.FloatTensor([3]) # 실수 텐서 생성자
print(f'm1.shape => {m1.shape}')
print(f'm2.shape => {m2.shape}')
print()

# 덧셈 연산
m3 = m1+m2
print(f'm3 => {m3}\nm3.shape=> {m3.shape}') # m1의 각각의 값에 3을 더함.
print()

m4 = m2+m1
print(f'm4 => {m4}\nm4.shape=> {m4.shape}') # m1의 각각의 값에 3을 더함.
```

```python

m1 = torch.FloatTensor([[1,2]])
m2 = torch.FloatTensor([[3], [4]])
print(f'm1.shape => {m1.shape}')
print(f'm2.shape => {m2.shape}')
print()

# 행렬 곱셈 연산
m3 = m1*m2
m4 = m1.mul(m2)
m5 = m2.mul(m1)
print(f'm3.shape => {m3.shape}\n')
print(f'm3 =>\n{m3}\n')

print(f'm4.shape => {m4.shape}\n')
print(f'm4 =>\n{m4}\n')

print(f'm4.shape => {m4.shape}\n')
print(f'm4 =>\n{m4}')
```

- add()
- sub()
- mul()
- div()

```python
# 원소 단위 연산 수행 메서드들 
m1.add(m2), m1.sub(m2), m1.mul(m2), m1.div(m2)
```
