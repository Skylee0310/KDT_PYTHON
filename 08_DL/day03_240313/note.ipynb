## 퍼셉트론 
- 사람의 뇌 신경세포(뉴런)의 동작과정을 真似て 만든 수학적 모델
    - 뉴런 : 
        * 화학 및 전기적 신호를 처리하고 전달하느데 관여하는 인간의 뇌에서 서로 연결된 신경 세포.
        * 가지돌기(dendrite)는 다른 뉴런으로부터 정보를 받아 세포체( cell body)를 거쳐 축삭(axon)을 통해서 다른 뉴런으로 신호 전달.
- 활성함수 : 강하게 반응할 건지 1 / 약하게 반응할 건지 0 이진분류기
