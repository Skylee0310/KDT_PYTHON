### X, Y축 레이블 설정 및 눈금 설정.
- x, y 축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정 => plt.xticks(), plt.yticks(), plt.tick_param()

```python
#그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^')
plt.title("[테스트]", fontsize = 'xx-large', color = 'g')

#x,y축 레이블
plt.xlabel("값", loc = "right", fontsize = 'xx-large', color = 'b')
plt.ylabel("결과", loc = "top", fontsize = 'xx-large', color = 'r',  rotation = 'horizontal') #위쪽 정렬. / 폰트 크기 큶. / 색깔 red / 수평!


#x, y 축 눈금 설정.
#plt.xticks(ticks = [0, 2, 4, ])
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'g')
plt.yticks(color = "red")
```

### 범례 설정 
- 여러 개의 그래프를 표시하는 경우 그래프 별 식별을 위해 사용.

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^') #label= '입력값'을 안에 쓰면 같은 것만 나옴.
plt.title("[테스트]", fontsize = 'xx-large', color = 'g')

# x,y축 레이블
plt.xlabel("값", loc = "right", fontsize = 'xx-large', color = 'b')
plt.ylabel("결과", loc = "top", fontsize = 'xx-large', color = 'r',  rotation = 'horizontal') #위쪽 정렬. / 폰트 크기 큶. / 색깔 red / 수평!


# x, y 축 눈금 설정.
#plt.xticks(ticks = [0, 2, 4, ])
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'g')
plt.yticks(color = "red")

#범례 설정
plt.legend(["Red 쏘영", "Green 승연"], loc = 'lower right')

#그래프 안에 텍스트 출력
plt.text(2, 2000, "Happy")
```

### 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y, 출력테스트)

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^') #label= '입력값'을 안에 쓰면 같은 것만 나옴.
plt.title("[테스트]", fontsize = 'xx-large', color = 'g')

# x,y축 레이블
plt.xlabel("값", loc = "right", fontsize = 'xx-large', color = 'b')
plt.ylabel("결과", loc = "top", fontsize = 'xx-large', color = 'r',  rotation = 'horizontal') #위쪽 정렬. / 폰트 크기 큶. / 색깔 red / 수평!


# x, y 축 눈금 설정.
#plt.xticks(ticks = [0, 2, 4, ])
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'g')
plt.yticks(color = "red")


plt.legend(["Red 쏘영", "Green 승연"], loc = 'lower right')

#그래프 안에 텍스트 출력
# plt.text(2, 2000, "쏘영승연", backgroundcolor='yellow', color = 'green')

# box={'facecolor':'.75', 'edgecolor':'k', 'boxstyle':'round', 'alpha':'0.2'}
# plt.text(2, 4000, "쏘영승연", bbox=box)

plt.text(2, 2000, 'text_add', backgroundcolor = 'yellow', color = 'red')

box = {'facecolor':'.75', 'edgecolor':'k', 'boxstyle':'round', 'alpha':0.3}
plt.text(2, 4000, 'text_add', bbox=box)
```

### 화살표 설정
- 그래프의 특정 지점에 화살표 표기
- text와 함꼐 종종 사용됨 ==>
- plt.annotate()

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^') #label= '입력값'을 안에 쓰면 같은 것만 나옴.
plt.title("[테스트]", fontsize = 'xx-large', color = 'g')

# x,y축 레이블
plt.xlabel("값", loc = "right", fontsize = 'xx-large', color = 'b')
plt.ylabel("결과", loc = "top", fontsize = 'xx-large', color = 'r',  rotation = 'horizontal') #위쪽 정렬. / 폰트 크기 큶. / 색깔 red / 수평!


# x, y 축 눈금 설정.
#plt.xticks(ticks = [0, 2, 4, ])
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'g')
plt.yticks(color = "red")

#범례 설정
plt.legend(["Red 쏘영", "Green 승연"], loc = 'lower right')

#화살표 설정
plt.annotate()
```
