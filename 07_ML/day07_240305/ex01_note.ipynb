### 이미지 데이터 분류
- 데이터 : sklearn.datasets의 MNIST => 디지털 숫자 0 ~ 9 손글씨 데이터셋.
- 피  처 : 이미지 픽셀 수
- 타  겟 : 0 ~ 9, 종 10개

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩 
from sklearn.datasets import fetch_openml

# 데이터 로딩
mnist = fetch_openml('mnist_784')
```

```python
# shape로 데이터와 타겟 데이터셋의 크기를 확인

mnist['data'].shape, mnist['target'].shape
```

```python
# describe()로 행 개수, 평균값 등의 대략적인 정보 확인.
mnist.data.describe()
```

[2] 데이터 전처리 <hr>
- 이미지 픽셀값 0 ~ 1로 스케일링

- featureDF를 생성한다.
- featureDF는 0 ~ 255 사이의 숫자인데 이것을 255로 나눔으로써 0과 1 사이의 수로 만들어 값을 정규화 하는 것이다.
- 정규화를 하는 이유는 모델 학습 속도를 향상시키고 특성 간 스케일 차이를 좁혀서 모델 성능을 향상시키며, 모델의 일반화를 더 개선하기 위함이다.

```python
# featureDF 생성
featureDF = mnist.data
featureDF = featureDF/255 # 0~ 255 사이였던 값을 0~1 사이로 변환
featureDF.describe()
```

- targetSR 생성 후 str타입이었던 숫자데이터들을 int로 형변환.
* -> 왜???? 
    - = 그 이유는 머신러닝 모델이 주로 수치형 데이터를 다루기 때문 + 데이터 형식 일치!

```python
# targetSR 생성
targetSR = mnist.target

targetSR.unique()
```

```python
# targetSR를 형변환한 뒤 unique()로 고유값 확인.
targetSR = targetSR.astype('int')
targetSR.unique()
```

[3] 기계 학습용 데이터셋 준비<hr>
- 학습용, 검증용, 테스트용 3개의 데이터셋 준비.
- 이 데이터의 경우에는 학습용 데이터 6만개와 테스트용 데이터 1만개로 분리.
- 검증용 데이터는 train_test_split을 이용해서 분리.

+ 데이터를 분리한 후에는 shape로 한번 더 확인을 해보자. 

```python
X_train = featureDF[:60000]
y_train = targetSR[:60000]
X_test = featureDF[60000:]
y_test = targetSR[60000:]

print(f'X_train : {X_train.shape}, y_train : {y_train.shape}')
print(f'X_test : {X_test.shape}, y_test : {y_test.shape}')
```

```python
# 모듈 로딩
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, 
                                                  test_size = 0.3,
                                                   stratify = y_train,
                                                   random_state = 5)
```

```python
# 데이터셋을 한번 더 확인하기.

print(f'X_train : {X_train.shape}, y_train : {y_train.shape}')
print(f'X_val : {X_val.shape}, y_val : {y_val.shape}')
```

[4-2] 확률적경사하강법(SGD) 기반 학습<hr>
- 대량의 데이터를 학습 시 랜덤으로 데이터셋 1개를 추출해서 학습 진행 => 최적화
- 장점 : 속도 증가
- 단점 : 정확도 감소

```python
# 모듈 생성
from sklearn.linear_model import SGDClassifier

# 인스턴스 생성
lr_sgd = SGDClassifier(loss='log_loss', random_state = 5) # log_loss : 실제 값을 예측하는 확률에 로그를 취하여 부호를 반전시킨 값

# 학습
lr_sgd.fit(X_train, y_train)

# 모델 파라미터 확인 -> 모델 파라미터 = 모델이 학습하는 동안 자동으로 조절되는 값.
print(lr_sgd.n_iter_) # 몇번 학습했는가
```

```python
# 각각의 점수를 변수에 대입해서 출력한다.
```

### [coef] 
- 회귀 모델에서 각 독립 변수의 계수(coefficients)
- 개수 : 회귀 모델에 사용된 독립 변수의 개수와 동일
- 해당 독립 변수가 종속 변수에 미치는 영향을 나타냅니다.

### [intercept_]
- 회귀 모델에서의 절편(intercept) 
- 개수 : 1개(모든 독립 변수가 0일 때의 종속 변수의 예측값) 
- 모델의 기본적인 위치를 나타내며, 모든 독립 변수가 0이 되었을 때의 종속 변수의 기댓값을 의미합니다.

```python
#


print(f'[Before ] lr_sgd.coef_ ==> \n{lr_sgd.coef_.shape}\n') # 독립변수의 개수와 같다(=피처)
print(f'[Before ] lr_sgd.intercept_ ==> \n{lr_sgd.intercept_}')
```

- X: 독립 변수(피처)로 구성된 배열 또는 행렬입니다. 주로 훈련 데이터의 피처들을 나타냅니다.

- y: 종속 변수(타겟)로 구성된 배열입니다. 주로 훈련 데이터의 타겟 값을 나타냅니다. 일부 모델은 이 인자를 필요로 하지 않을 수 있습니다.

- classes: 분류 모델에서 사용되며, 가능한 클래스의 목록을 제공합니다. 이는 다중 클래스 분류 문제에서 클래스 레이블의 유일한 값들을 정의하는 데 사용됩니다.

- sample_weight: 각 샘플에 대한 가중치를 나타내는 배열입니다. 샘플의 상대적 중요도를 조절하는 데 사용됩니다.

- check_input: 입력 데이터의 유효성을 확인할지 여부를 결정하는 매개변수입니다. 기본적으로 True로 설정되어 있습니다.

- X_idx_sorted: DecisionTreeClassifier나 DecisionTreeRegressor와 같은 결정 트리 모델에서 사용됩니다. 입력 데이터의 특성에 대한 인덱스를 정렬한 배열을 제공합니다.

- y_idx_sorted: X_idx_sorted와 유사하지만 타겟 변수에 대한 인덱스를 정렬한 배열을 제공합니다.

- fit_params: 일부 추정기에서 추가적인 매개변수를 설정하는 데 사용됩니다.
