### Decision Tree 결정 트리

- Y/N 질문(특정 기준)으로 학습 진행 => 스무고개 퀴즈
- 질문(특정 기준)을 뭘로 하느냐에 따라 성능이 크게 달라짐.
- 특정 기준에 따라 데이터 구분 => 의사결정 나무 모델
- 직관적, 범용성, 해석력이 좋지만 데이터에 민감함
- 데이터 사전 가공에 대한 영향이 매우 적음
- 분류와 회귀 모두 가능한 지도 학습 모델 중 하나

- 고려사항 : 
    * 불순도(impurity) : 서로 다른 데이터가 얼마나 섞여 있는지
    * 정보 이득 : 나누고 나서 불순도의 차이.
