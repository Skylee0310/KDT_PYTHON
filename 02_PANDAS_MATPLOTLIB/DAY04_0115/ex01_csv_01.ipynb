## [ csv 데이터 파일 다루기 ]

(1) 모듈 로딩 <hr>

```python
import pandas as pd
```

(2) 데이터 준비 <hr>

- 파일 이름 => 경로 + 파일명
- 경로 => 2가지 종류
     * 절대 경로 : 드라이브부터 파일까지의 전체 경로(ex : C:\User\data\a.txt)
     * 상대 경로 : 현재 파일을 기준으로 경로 설정
        - ./ : 현재 위치 (ex : ./a.csv)
          ===> 현재 파일과 동일한 폴더, 위치에 존재.
        - ../ : 현재 위치에서 한 단계 위로 올라간 위치 (ex: ../Data/b.csv)
          ==> 현재 파일이 존재하는 폴더 밖에 존재하는 파일.

(5) read_csv 파일 매개변수 살펴보기 <hr>

- header : DataFrame의 컬럼명을 설정하는 파라미터.
    * 기본값 : csv 파일의 첫번째 라인의 데이터를 컬럼명으로 설정.
    * None : csv 파일의 첫번째 라인이 컬럼명이 아님을 의미 ( = 컬럼명 미지정)
    * 행번호 : csv 파일의 행번호 라인이 컬럼명으로 설정, 그 이전 행번호 데이터는 소실.

```python
pd.read_csv(file2, header=None) #열 이름이 자료가 됨.
```

```python
pd.read_csv(file2, header=1) #1열이 열 이름 인덱스가 됨.
```

- index_col : 열(Column) 중에서 행 인덱스로 설정, set_index()와 동일한 역할 수행.

```python
#pd.read_csv(file2, index_col = ['Rank', 'Title']) #멀티 인덱스
pd.read_csv(file2, index_col = 'Rank') #단일한 열을 인덱스로 사용.
```

- sep/ delimiter : CSV 파일에서 데이터 구분자 설정.
    * 기본값 : 쉼표(,)

```python
#pd.read_csv(file, header = None, sep=';') # 구분자 기본값을 쉼표에서 세미콜론으로 변경.
```

- skiprows : csv 파일에서 읽어 들이지 않을 행 번호 지정. => 앞부분
- skipfooter : csv 파일에서 읽어 들이지 않을 행 번호 지정. => 끝부분

```python
pd.read_csv(file, header = None, sep=';', skiprows=3) #3줄을 읽어오지 않음.
```

```python
pd.read_csv(file, header = None, sep=';', skiprows=3, skipfooter=2, engine="python") # 마지막 두 줄은 출력하지 않음.
```

- usecols : 컬럼에서 특정 열(Column)만 선택해서 로딩함.
    * usecols = [0,2, ...]
    * usecols = ['컬럼명1', '컬럼명2'...컬럼명n']

```python
pd.read_csv(file, sep = ';', skiprows = 3, header=None, usecols = [0, 2])
```

```python
pd.read_csv(file2, usecols = ['Rank', 'Year'])
#pd.read_csv(file2, usecols = [0, 4])
```
